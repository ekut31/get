```python
import numpy as np
import matplotlib.pyplot as plt
from textwrap import wrap

x = np.array([9.966, 12.5, 15.095])
y = np.array([26.51, 31, 34.67])
t = np.polyfit(x, y, 1)
f = np.poly1d(t)
print(f)

fig, ax = plt.subplots()

ax.minorticks_on()
ax.grid(which='minor',color='gray',linestyle=':')

ax.plot(x, y, color='orange', linestyle='-', linewidth= 1, marker='o', markersize=3, markerfacecolor='white', markeredgewidth=1.5, markeredgecolor='orange')
ax.plot(x, f(x), color='red', linestyle='-', linewidth= 1, marker='o', markersize=3, markerfacecolor='white', markeredgewidth=1.5, markeredgecolor='red', label=f)

ax.set_xlim(np.min(x), np.max(x) + 3)
ax.set_ylim(np.min(y), np.max(y) + 0.5)

ax.set_xlabel(', ')
ax.set_ylabel(' ')

ax.set_title("      ", wrap = True, loc = 'center')

ax.grid(True, color='lightgray', linestyle='--')

text_x = np.mean(x)
text_y = np.mean(y)
ax.text(text_x, text_y,  "",  fontsize=12, color='red', ha='center', va='center')

ax.legend()

plt.show()
fig.savefig('1.svg')
```
